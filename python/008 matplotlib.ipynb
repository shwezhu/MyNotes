# 1. Pyplot
----

### 1.1. 常用函数

- 指定数据和line的style: `plt.plot([)`, `plt.stem()`, 这两个函数参数一样, 只是样式不一样
- 指定x,y axis最大/小值: `plt.axis()`
- Shows 'label': `plt.legend()` 
- Shows the grid, default is False: `plt.grid(True)`

### 1.2. Controlling Line Properties

linestyle的值: `-`, `--`, `:`

marker的值: `o`, `^`, `*`, `+`

color的值: `b`, `g`, `y`, `r`, `c`, `k`, `w`

### 1.3. 综合例子

```
# Generate 30 evenly spaced points between 0 and 10
t = np.linspace(0, 10, 30)

# Create an array of the sin of every point in array t
x = np.sin(t)

# plot the graph of x against t
plt.figure(figsize=(10,5))
plt.xlabel('time(s)')
plt.ylabel('amplitude')
plt.title('Graph of Sin(t)')
# plt.plot(t, x, color='blue', marker='o' )
plt.stem(t, x,'red')
# 要加下面这行代码, 否则label不显示
plt.legend()
plt.grid(True)
plt.show()
```

# 3. Multiple Plots
---

````python
import matplotlib.pyplot as plt

# Define the Data
t = np.linspace(0, 10, 100)
x = np.sin(t)

# Initialize the Objects needed to plot the Graph
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(12, 15)) # 3x1 grid canvas

#######################
# Plot on first grid
#######################
# Set plot title and axes labels
ax[0].set(title = "Sinusodial Sine Wave",
       xlabel = "time in secs",
       ylabel = "sin(t)")

# Define x and y axes and plot
ax[0].plot(t, np.sin(t), color='red', marker = 'o', label='sine wave')
ax[0].legend() # shows the legend of the graph, it reflects the 'label' keyword argument in the plot function above
ax[0].grid(True) # shows the grid, default is False


#######################
# Plot on second grid
#######################
# Set plot title and axes labels
ax[1].set(title = "Sinusodial Cosine Wave",
       xlabel = "time in secs",
       ylabel = "cos(t)")

# Define x and y axes and plot
ax[1].plot(t, np.cos(t), color='blue', marker = 'x', label='cosine wave')
ax[1].legend() # shows the legend of the graph, it reflects the 'label' keyword argument in the plot function above
ax[1].grid(True) # shows the grid, default is False

#######################
# Plot on third grid
#######################
# Set plot title and axes labels
ax[2].set(title = " Discrete Time Signal",
       xlabel = "time in secs",
       ylabel = "cos(t)")

# Define x and y axes and plot
ax[2].stem(t, np.cos(t), 'yellow')
ax[2].legend() # shows the legend of the graph, it reflects the 'label' keyword argument in the plot function above
ax[2].grid(True) # shows the grid, default is False

# needed to explicitly show the graph (optional in an ipython notebook environment)
plt.show()
````
