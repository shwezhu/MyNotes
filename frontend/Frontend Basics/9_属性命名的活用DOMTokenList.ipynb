# 属性命名
---
看到一些命名, 很奇怪中间竟然有空格, 类似下面这样

```html
<div class="a b c"></div>
```

我最初的理解是创建了多个`div`, 后来才发现理解错了. 其实是相当于给一个`div`起了多个名字. 举个栗子:

```html
<tag class="a b">....</tag>
<tag class="a">....</tag>
<tag class="b">....</tag>
```

`.a.b {}` in your css will target the first tag.

`.a {}` will target the first and second tags.

`.b {}` will target the first and third tags.

然后TokenList呢指的就是class的两个名字`a`和`b`,然后这个list可以通过interface`Element`的`classList`property来获得.

具体例子可以参考下面这个文章: https://developer.mozilla.org/en-US/docs/Web/API/DOMTokenList


# 贪吃蛇的应用

---

先看HTML和CSS的代码:

```html
<div class='grid'>
	<div></div>
	<div></div>
	<div></div>
</div>
```

CSS:

```css
.grid div {
  width: 20px;
  height: 20px;
}

.snake {
  background-color: blue;
}
```

思路呢就是grid里面的div就是蛇的身体,有好多个,有选择的控制其中一些div显示出颜色,就相当于模拟出蛇的形状了. 比如让第一个div显示紫色,其他div都是无色,然后下一秒让第二个div显示紫色,把第一个div的颜色去掉, 这就让蛇移动了. 

那么问题来了怎么可以简单控制每个div的颜色呢? 你可能会注意到css里面的snake,但是html明明没有snake的class. 刚开始我也在奇怪, 看懂了tokenlist后才明白其中的妙用. 可以利用`Element.classList`来获得一个类的class名字列表, 然后通过给它添加名字`snake`, 再加上原本css里面的那句背景颜色, 就会自动应用到这个div上. 不想显示的时候再remove掉这个名字.

```js
const squares = document.querySelectorAll(".grid div");
// so the 3rd div in our grid being 2 (or the HEAD), and 0 being the end
let currentSnake = [2, 1, 0]; 

//to start, and restart the game
function startGame() {
      // 先清空之前蛇的身体
  currentSnake.forEach(index => squares[index].classList.remove("snake"));
  squares[appleIndex].classList.remove("apple");

  intervalTime = 500;
  currentSnake = [2, 1, 0];
  currentSnake.forEach(index => squares[index].classList.add("snake"));
  interval = setInterval(moveOutcomes, intervalTime);
}
```
