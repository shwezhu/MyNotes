# 2. DOM Tree 一切控制的基础
---

A DOM tree is a kind of tree whose `nodes` represent an HTML or XML document's contents. 

## 2.1. DOM Node

下面这就是个Tree, `Document`是根node,然后下面是`<html>`. `<html>` is a node in the document tree. It has 2 children: `<head>` and `<body>` nodes. `<body>` is also a node having 3 children: a comment `<!-- Page Body -->`, heading `<h2>`, and paragraph `<p>`. The parent of the `<body>` node is `<html>` node. The tags in the HTML document represent a node, what's interesting is that regular text is also a node. The paragraph node `<p>` has 1 child: the text node `"Thank you for visiting my web page!"`.

```html
<!DOCTYPE html>
<html>
  <head>
    <title>My Page</title>
  </head>
  <body>
    <!-- Page Body -->
    <h2>My Page</h2>
    <p id="content">Thank you for visiting my web page!</p>
  </body>
</html>
```

**When a web browser parses an HTML document, it builds a DOM tree and then uses it to display the document**.

### 2.1.1. Node Types

Nodes have types, the element type being one of them.

How can you distinguish these different types of nodes? The answer lays in the **DOM `Node` interface**, particularly in the **`Node.nodeType` property**.

`Node.nodeType` can have one of the following values that represents the type of the node:

+ Node.ELEMENT_NODE

+ Node.ATTRIBUTE_NODE

+ Node.TEXT_NODE  上面的那段内容,就是字符串它也是node,他的类型就是这个

+ Node.CDATA_SECTION_NODE

+ Node.PROCESSING_INSTRUCTION_NODE

+ Node.COMMENT_NODE  上面那个注释就是这个类型


## 2.2. DOM Element

DOM Tree里面最重要的两个概念就是`node`和`element`了. node根据上面的例子可以看出,多个nodes组成了一个DOM tree.

那什么是element?

可以这么说, element属于node,但并不是所有的node都是element, 比如上面例子中的那段text,就是一个字符串,还有注释(comment)这都是node,但不是element. element是用tag定义的,比如`<html>`, `<div>`, `<p>`, `<button>`这都是element.

# 3. Document 
---

## Method

#### Document.getElementById()

Returns **an** object reference to the identified element.

#### Document.getElementsByClassName()

Returns **a list of** elements with the given class name.


#### Document.querySelectorAll()

Returns **a list of** all the Element nodes within the document that match the specified selectors.

## Events

Listen to these events using `addEventListener()` or by assigning an event listener to the `oneventname` property of this interface.

#### `beforescriptexecute` 

Fired when a static `<script>` is about to start executing.

#### `scroll`

Fired when the document view or an element has been scrolled.

## Properties

This interface also inherits from the `Node` and `EventTarget` interfaces.

#### `Document.body`

Returns the `<body>` or `<frameset>` node of the current document.

#### `Document.children` Read only

Returns the child elements of the current document.
