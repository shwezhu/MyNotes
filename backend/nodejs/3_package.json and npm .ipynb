# Init NodeJS Project

---

初始化, 完成后自动创建一个文件`package.json`. 

```shell
npm init
```

利用npm安装module, 运行后, module会被自动安装到 `./node_modules`, 该目录下的modules, 都可以通过builtin函数`require`来进行使用. 

```shell
npm install xxx
```

## Example

首先, `npm init`运行之后, 文件夹下只产生一个文件即`package.js`, 内容如:

```json
{
  "name": "package_name",
  "version": "1.0.0",
  "description": "my description",
  "main": "app.js",
  "scripts": {
    "test": "echo \"Error: no test specified\" && exit 1"
  },
  "author": "",
  "license": "ISC"
}
```

运行`npm install async`安装 async module, 此时文件夹中又多了两个文件 `node_modules` `package-lock.json`:

```shell
ls
node_modules      package-lock.json package.json

ls node_modules 
async

cat package.json 
{
  "name": "package_name",
  "version": "1.0.0",
  "description": "my description",
  "main": "app.js",
  "scripts": {
    "test": "echo \"Error: no test specified\" && exit 1"
  },
  "author": "",
  "license": "ISC",
  "dependencies": {
    "async": "^3.2.4"
  }
}

cat package-lock.json 
{
  "name": "package_name",
  "version": "1.0.0",
  "lockfileVersion": 2,
  "requires": true,
  "packages": {
    "": {
      "name": "package_name",
      "version": "1.0.0",
      "license": "ISC",
      "dependencies": {
        "async": "^3.2.4"
      }
    },
    "node_modules/async": {
      "version": "3.2.4",
      "resolved": "https://registry.npmjs.org/async/-/async-3.2.4.tgz",
      "integrity": "sha512-iAB+JbDEGXhyIUavoDl9WP/Jj106Kz9DEn1DPgYw5ruDn0e3Wgi3sKFm55sASdGBNOQB8F59d9qQ7deqrHA8wQ=="
    }
  },
  "dependencies": {
    "async": {
      "version": "3.2.4",
      "resolved": "https://registry.npmjs.org/async/-/async-3.2.4.tgz",
      "integrity": "sha512-iAB+JbDEGXhyIUavoDl9WP/Jj106Kz9DEn1DPgYw5ruDn0e3Wgi3sKFm55sASdGBNOQB8F59d9qQ7deqrHA8wQ=="
    }
  }
}
```

然后使用npm安装`express` module, 之后查看`node_modules`文件夹, 一下子多出了几十个modules, 这是为什么呢, 有个东西叫依赖dependence, 就是`express` module的实现依赖了很多其他的modules. 然后再看`package-lock.json`文件, 多出了好多行, 都是每个modules的版本信息, 再看`package.json`只是在dependencies的async下面多了一个express. 所以可以看出`package-lock.json`记录了所有在文件夹`node_modules`下的modules, 而`package.json`只是简单记录了此项目的直接依赖. `package-lock.json` 的作用很简单, 就是形成了一个依赖树,记录每个module的精确版本,而如果只有`package.json`, 那也就只能保证`expree`和`async`两个模块的版本不变, 而间接依赖就未必了(其他人用此项目的时候, 显然git版本控制的时候不用track`node_modules`, 只用track这两个json文件就行.)
