# 安装

---

安装教程:https://www.mongodb.com/docs/manual/tutorial/install-mongodb-on-os-x/

如果上面的网站进不去, 进这个找: https://www.mongodb.com/docs/manual/installation/

安装并运行:
```shell
brew tap mongodb/brew

brew update

brew install mongodb-community@6.0

brew services start mongodb-community@6.0

brew services stop mongodb-community@6.0

# 查看
brew services list

# 运行
mongosh
```

配置文件所在位置(都在上面的教程中可以找到):
```shell
configuration file: /opt/homebrew/etc/mongod.conf
log directory: /opt/homebrew/var/log/mongodb
data directory: /opt/homebrew/var/mongodb
```

查看配置文件内容
```zsh
cat /opt/homebrew/etc/mongod.conf

systemLog:
  destination: file
  path: /opt/homebrew/var/log/mongodb/mongo.log
  logAppend: true
storage:
  dbPath: /opt/homebrew/var/mongodb
net:
  bindIp: 127.0.0.1, ::1
  ipv6: true
```

# 基本指令

---

```zsh
# 运行MongoDB shell
mongosh

# 查看数据库
test> show dbs
admin   40.00 KiB
config  72.00 KiB
local   40.00 KiB

# 创建数据库
# MongoDB only creates the database when you first store data in that database. 
test> use greenhouse
switched to db greenhouse
greenhouse>

# 创建名为temperature的表(collection) 
greenhouse> var date = new Date(2022, 09, 18, 22, 59);

greenhouse> db.temperature.insertOne( {value: 25.6, date: date} );
{
  acknowledged: true,
  insertedId: ObjectId("63273297b128d4558c64de09")
}

# 查看表
greenhouse> show collections
temperature

# 查找数据
db.temperature.find( {value: 25.6} )
[
  {
    _id: ObjectId("63273297b128d4558c64de09"),
    value: 25.6,
    date: ISODate("2022-10-18T14:59:00.000Z")
  }
]
```
