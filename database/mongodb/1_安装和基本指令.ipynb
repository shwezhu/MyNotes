# 安装

---

安装教程:https://www.mongodb.com/docs/manual/tutorial/install-mongodb-on-os-x/

如果上面的网站进不去, 进这个找: https://www.mongodb.com/docs/manual/installation/

安装并运行:
```shell
brew tap mongodb/brew

brew update

brew install mongodb-community@6.0

brew services start mongodb-community@6.0

brew services stop mongodb-community@6.0

# 查看
brew services list

# 运行
mongosh
```

配置文件所在位置(都在上面的教程中可以找到):
```shell
configuration file: /opt/homebrew/etc/mongod.conf
log directory: /opt/homebrew/var/log/mongodb
data directory: /opt/homebrew/var/mongodb
```

查看配置文件内容
```zsh
cat /opt/homebrew/etc/mongod.conf

systemLog:
  destination: file
  path: /opt/homebrew/var/log/mongodb/mongo.log
  logAppend: true
storage:
  dbPath: /opt/homebrew/var/mongodb
net:
  bindIp: 127.0.0.1, ::1
  ipv6: true
```

# 使用MongoDB Shell的流程

---

- **运行MongoDB shell**

```shell
mongosh

show dbs
```

- **创建数据库`greenhouse`**

```shell
# MongoDB only creates the database when you first store data in that database. 
use greenhouse
```

- **创建collection`users`**

```shell
db.createCollection("users")
{ ok: 1 }
```

- **查看此db中的所有collections**

```shell
db.getCollectionNames()
[ 'users', 'humidity', 'temperature' ]
```

- **向collection`users`插入一个document**

```shell
db.users.insert({ age: 20, create_date: new Date() })

# 查询
db.users.find()
```

- **删除collection`users`**

```shell
db.users.drop()
```

- **删除db`greenhouse`**

```shell
db.dropDatabase()
```

- **删除数据库`shcool`**

```shell
mongosh

use school

db.dropDatabase()
```

- **删除db`school`中的collection`users`**

```shell
mongosh

use school

db.users.drop()
```
