# 1. JVM Architecture
---

JVM is only a specification, and its implementation is different from vendor to vendor. For now, let’s understand the commonly-accepted architecture of JVM as defined in the specification.



JVM 包括两部分 Classloader, Execution Engine, JVM开始工作的第一阶段(上图loading部分)就是通过classloader**把类(bytecode)导入到内存ram中**, 程序运行不就得先从磁盘进入内存吗. 每个java类或者接口一般都是单独在一个`.java`文件, 所以JVM会导入多个`.class`文件, 但有个主文件, 就是那个含有`static main()`, 也就是说classloader会先load含有这个静态方法的字节码文件. 之后再导入其他类(**loading的过程是dynamic**, 用到谁导入谁, 不是一下全部导入). 

当然导入类的时候也会遵守一些原则, 具体在上一节, 本章只专注于内存结构.
