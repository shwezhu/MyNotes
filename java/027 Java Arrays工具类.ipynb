`java.util.Arrays`, 常见的数组操作，这个类都提供了静态方法可供直接调用:

- 创建数组
- 数组排序
- 数组检索
- 打印数组
- setAll
- parallelPrefix
- 数组转 List
- 数组转流

----

## 1) 创建数组


```java
public class Main {
    public static void main(String[] args) {
        String[] intro = new String[] { "a", "b", "c", "d" };
        String[] intro_1 = Arrays.copyOf(intro, 2);
        String[] intro_2 = Arrays.copyOfRange(intro, 1, 3);
        System.out.println(Arrays.toString(intro_1));
        System.out.println(Arrays.toString(intro_2));
    }
}

[a, b]
[b, c]
```

## 2) 数组排序

```java
String[] intro = new String[] { "e", "b", "hh", "d" };
String[] intro_1 = Arrays.copyOf(intro, 4);
Arrays.sort(intro_1);
System.out.println(Arrays.toString(intro_1));
```

## 3) 数组检索

数组排序后就可以使用 Arrays 类的 `binarySearch()` 方法进行二分查找了 

```java
public class Main {
    public static void main(String[] args) {
        String[] intro = new String[] { "e", "b", "hhd", "d" };
        Arrays.sort(intro);
        int index = Arrays.binarySearch(intro, "d");
        System.out.println(index);
    }
}
```


## 4) 打印数组

数组是对象，直接 System.out.println 的话，结果是这样的：

```java
[Ljava.lang.String;@3d075dc0
```

最优雅的打印方式，是使用 `Arrays.toString()`.


## 5) setAll()

Java 8 新增了 `setAll()` 方法，它提供了一个函数式编程的入口:

```java
public class Main {
    public static void main(String[] args) {
        String[] intro = new String[] { "e", "b", "hhd", "d" };
        // lambda 一行代码的时候 可以省略return
        Arrays.setAll(intro, i -> intro[i] + "a" );
        System.out.println(Arrays.toString(intro));
    }
}
```

## 6) parallelPrefix

`parallelPrefix()` 方法和 `setAll()` 方法一样，也是 Java 8 之后提供的，提供了一个函数式编程的入口，通过遍历数组中的元素，将当前下标位置上的元素与它之前下标的元素进行操作，然后将操作后的结果覆盖当前下标位置上的元素。

```java
int[] arr = new int[] { 1, 2, 3, 4};
// 当前元素和前一个元素, 所以需要两个参数
Arrays.parallelPrefix(arr, (left, right) -> left + right);
System.out.println(Arrays.toString(arr));
```
