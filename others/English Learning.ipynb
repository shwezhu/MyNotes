### overweight
+ adjective

> Now Frolic was middle-aged, `overweight`, lazy, and no longer needing to chew.      
Lower-income women are more likely than other women to be `overweight`, though the pattern among men may be the opposite.       
Being `overweight` or obese are both terms for having more `body fat` than what is considered healthy. However, the term "obese" generally means a much higher amount of `body fat` than "overweight."       

### fat
+ noun     
Fat is a form of body tissue on animals. It’s a squishy layer that stores energy and protects organs. 
+ adjective    
Fat is also an adjective meaning chubby or thick, as in a fat piglet or a fat wallet full of money.

### chubby
+ adjective     
Someone who's chubby is plump or `slightly overweight`. If the vet mentions that your cat's getting a little chubby, you might need to cut down on the cat treats.

###  Spoken English
---

+ **Hang in there!**          
remain persistent and determined in difficult circumstances.        
> "in the second half, we just had to hang in there"

+ **keep going!**        
make an effort to live normally when in a difficult situation.       
> she had to `keep going` for the sake of her boys    

+ **go for it!**              
If you go for it, you make a big effort to achieve something or you decide to do something. When you set the right goals for yourself, you will feel ready and willing to go for it.           
> Don't throw away your chances — just go for it!

+ **Are you…?** is used in the following situations:

To ask someone about his identity. For example, Are you a vegetarian/Jewish/an artist?

To ask someone what he’s doing at the present moment. For example, Are you writing/dreaming/typing?

To ask someone about his identity (adjective). For example, Are you mad/serious/sick?

+ **Do you…?** is used in the following situations:

To ask someone what he generally does. For example, Do you smoke/drink/dance?

To ask about the ownership/possession. For example, Do you have a cat?/Do you own that house?

I think it’s best to give you a few examples:

> Do you like swimming? Are you a swimmer?          
Do you play tennis? Are you a tennis player?           
Do you climb mountains? Are you a mountain climber?             
Do you feel okay? Are you okay?                  

+ **not really**      

—used to say "no" in a way that is not very forceful or definite

> "Was the movie good?" "Not really."             
"Do you want to go to a movie?" "No, not really."


### understand/noted

An example:
Someone may say "can you fold your clothes like this instead in the future?" And you can respond "noted" to let them know that you will try to remember to do that in he future. For understood/understand, someone can teach you how to do a math problem and once you figure out how they are doing it, you now understand the problem.

We also use understand when responding to people being unable to do something. For example, if someone says "I'm sorry I can't help you today, I'm sick" we would respond with "I understand" to let them know that it is okay and we don't mind.

**respond politely**

Often the best way to indicate that you have understood is to summarize in one sentence what your understanding will lead you to do next.
e.g., 

> "I understand. I will get back to you with the revisions you requested by Monday."

> "I see. I will make sure to include or discuss your idea that .... in my next paper."

or

> I really appreciated your explanations in the last session I attended, they helped me to grasp/understand [topic] and now I feel much more confident. Thank you.

> Thank you so much for your clear and unambiguous explanations on [topic]. They have helped me tremendously.
