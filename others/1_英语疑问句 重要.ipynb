# 助动词

---
先说一下, 有助动词就会有主动词, 一个是副一个是主. 助动词, 助动词, 就精确在这个助字上, 即帮助动词来完成一些特定的目的. 举个例子:

She does not eat fish.

1. does 就是助动词； eat 就是主动词。

2. Does eat 合在一起就是**谓语动词**。


An auxiliary verb (or a **helping verb** as it's also called) is used with a main verb to help express the main verb's tense, mood, or voice.

## Examples of Auxiliary Verbs Expressing Tense


I `will` walk to work.

She `was` waiting for an hour. 

She `is` waiting in the hall.

She `will be` waiting outside.

She `had` drunk it before we arrived.

She `has` drunk it already.

## Examples of Auxiliary Verbs Expressing Voice

Our dessert `was` eaten by the dog.

The geese `are` driven through the snicket.

The phone `will be` disconnected tomorrow.

## Examples of Auxiliary Verbs Expressing Mood

`Did` you win?
(Here, the auxiliary verb to do is used to form the interrogative mood, i.e., to ask a question.)

`Don't` forget your wallet.
(Here, the auxiliary verb to do (in its negative form) is used to form the imperative mood, i.e., to give an order.)

# 疑问句

---

助动词有很多, have, will, must, may, should, be, do, 这都是助动词, 都可以用到疑问句中, 每个助动词都有自己的时态和形式别忘了, 比如 do, 你提问的时候可以问 Do you, did you, does he, 等等


## 特殊疑问句

疑问词放至句首，**（找be, can, would, will, 等助动词）抄在疑问词后，没有则用do，does**，改（出现some，要考虑是否改为any。出现I 改为 you, 出现 we 改为 you ,出现my 改为your,出现our改为your。最后是加问号

例子:

I usually get up at six forty .--- `When` `do` you usually get up?　没找到助动词,就用do/does代替, 因为是问什么时候, 所以用when.

I can see five birds in the tree.---`How many` birds can you see?　 问多少, 所以选how many, 找到了助动词can,所以不用加do/does. 但是这个还特殊, 把birds提前了, 然后再放 can, 这种类型的都是这么问, 

比如, There are three apples. ---`How many` apples are here? 因为are(be动词)也属于助动词. 

That is my shirt . ---`Whose` shirt is that?　找到主动词is, 因此不用do/does

They are five yuan. -- `How much` are they?

I like math best. --`What subject` `do` you like best? 这句就比较像上面的how many的例子, 疑问词后面不是直接跟助动词而是要问的东西,自己体会语感.

They bought a new bike yesterday. --`What` `did` they buy yesterday?

1.“事或物”，特殊疑问词用what Eg:

2、 划线部分是“人”，特殊疑问词用who　 

Eg:She is my sister.  ---`Who` is she ?　

3、 划线部分是“地点”，特殊疑问词用where 

Eg:The apple is on the desk.---`Where` is the apple ?　　

４、划线部分是“时间”，特殊疑问词用what time或when Eg:

It's six thirty . ---`What time` is it ?　　

I usually get up at six forty .--- `When` `do` you usually get up?　　

5、划线部分是“年龄”，特殊疑问词用how old Eg:

I am twelve . ---How old are you?　　　

My mother is thirty- two . ---`How old` is your mother ?　　

6、划线部分是“职业”，特殊疑问词用what　 Eg:

Tom is a worker.  ---`What` is Tom?/What do you do ?　　

7、划线部分是“颜色”，特殊疑问词用what colour　 Eg:

My hat is blue . ---`What colour` is your hat？　　

８、划线部分是“数量”，特殊疑问词用how many或how much　 Eg:

I can see five birds in the tree.---`How many` birds can you see ?　　　

There is some tea in the cup.---`How much` tea is there in the cup ?　　

９、划线部分是“多少钱”，特殊疑问词用how much　　Eg:

This book is ten yuan . ---`How much` is this book ?　　

１０、划线部分是“形容词性物主代词或名词性物主代词”，特殊疑问词用whose Eg:

That is my shirt . ---`Whose` shirt is that ?
